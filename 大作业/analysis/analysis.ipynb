```python
import numpy as np
import time
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib.font_manager import FontProperties
import seaborn as sns
import os

#路径需要修改
data = pd.read_csv('all_final.csv', encoding="utf-8")
print(data['薪资_平均值'])
```

```python
#工资分布
plt.boxplot(data['薪资_最小值'],meanline=True,patch_artist='blue')
plt.ylabel(u'薪资_最小值（人民币万元/年）')
plt.ylim(0,1000000)
plt.show()

plt.boxplot(data['薪资_最大值'],meanline=True,patch_artist='blue')
plt.ylabel(u'薪资_最大值（人民币万元/年）')
plt.ylim(0,1000000);
plt.show()

plt.boxplot(data['薪资_最大值']-data['薪资_最小值'],meanline=True,patch_artist='blue')
plt.ylabel(u'薪资_上下限差（人民币万元/年）')
plt.ylim(0,500000);
plt.show()
```

```python
y = [0]*8
for i in range(1,9):
    y[i-1] = np.mean(data[data.学历要求 == i]['薪资_平均值'])
print(y)
plt.bar(['初中及以下','中技','中专','高中','大专','本科','硕士','博士'], y)
plt.xlabel(u'学历要求')
plt.ylabel(u'频数')
plt.title(u'平均薪资相对于学历要求的分布柱状图')
plt.show()
```

```python
important = data[['公司规模','学历要求','工作经验','薪资_平均值']]
pearson = important.corr()
kendall = important.corr('kendall')
spearman = important.corr('spearman')
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus'] = False 
plt.plot(figsize = (6,6))  

sns.heatmap(pearson, annot=True, vmax=1,vmin = -1, square=True)
```
